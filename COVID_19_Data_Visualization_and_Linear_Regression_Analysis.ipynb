```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lm=LinearRegression()
```

```python
msk=np.random.rand(len(df))<0.8
train=df[msk]
test=df[~msk]
```

```python
train_x=np.asanyarray(train[["Deaths"]])
train_y=np.asanyarray(train[["Confirmed"]])
lm.fit(train_x,train_y)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.predict(train_x)
```

```python
plt.scatter(train["Deaths"],train["Confirmed"])
plt.plot(train_x,lm.coef_[0][0]*train_x,lm.intercept_[0],"-r",color="deeppink")
plt.xlabel("Deaths cases")
plt.ylabel("Confirmed cases")
plt.title("Covid Deaths")
```

```python
test_x=np.asanyarray(test[["Deaths"]])
test_y=np.asanyarray(test[["Confirmed"]])
test_y=lm.predict(test_x)
```

```python
print("Mean Absolute Error:%.2f"%np.mean(np.absolute(test_y-test_y)))
```

```python
print("Residual sum of_ square(MSE):%.2f"%np.mean(np.absolute(test_y-test_y)**2))
```

```python
print("R2_score:%.2f"%r2_score(test_y,test_y))
```

```python
plt.scatter(test["Deaths"],test["Confirmed"])
plt.plot(test_x,lm.coef_[0][0]*test_x,lm.intercept_[0],"-r",color="deeppink")
plt.xlabel("Deaths cases")
plt.ylabel("Confirmed cases")
plt.title("Covid Deaths")
```
